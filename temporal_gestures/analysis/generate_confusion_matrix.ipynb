## Generate Confusion Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt    
from sklearn.metrics import confusion_matrix 

GESTURE_NAMES = ["FFS", "SFF", "FPFF", "FFPF", "Stable", "Speedup", "Slowdown", "Square"]
cm = confusion_matrix(expected, actual, labels =GESTURE_NAMES)

sns.set_theme(rc={"figure.figsize": (15, 12)})

ax= plt.subplot()
fig = sns.heatmap(cm, annot=True, fmt='g', ax=ax);  # 'g' disables scientific notation




# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(GESTURE_NAMES); ax.yaxis.set_ticklabels(GESTURE_NAMES)

plt.savefig("./confusion_matrix.png")
```
